## 1. Introduction

In this lab, we enhance the AEP (American Electric Power) energy consumption dataset by incorporating holiday information. This helps in analyzing how energy usage patterns change during holidays.

```python
import pandas as pd
import numpy as np
```

## 2. Loading the Dataset

```python
df = pd.read_csv(r'C:\Users\arif\OneDrive - Higher Education Commission\Teaching\Courses\machine learning\Spring 25\Lab for students\3_Outlier_Identified.csv', parse_dates=['Datetime'])
df.head()
```

## 3. Extracting Date from Datetime

```python
# Extract only the date part
df['Date'] = df['Datetime'].dt.normalize()
df.head()
```

```python
print("Total records:", len(df))
```

## 4. Loading Holiday Data

```python
holiday = pd.read_csv(r'C:\Users\arif\OneDrive - Higher Education Commission\Teaching\Courses\machine learning\Spring 25\Lab for students\processed_holiday.csv', parse_dates=['Date'])
holiday.head()
```

```python
holiday.info()
```

## 5. Merging Holiday Information

```python
# Keep only Date and Holiday columns
holiday = holiday[['Date', 'Holiday']]

# Merge with main dataset
mergedf = pd.merge(df, holiday, on='Date', how='left')
mergedf.head()
```

## 6. Creating Binary Holiday Feature

```python
# Convert Holiday to binary (1 = Holiday, 0 = Normal Day)
mergedf['Holiday'] = mergedf['Holiday'].notnull().astype(int)
mergedf.head()
```

## 7. Saving the Final Dataset

```python
mergedf.to_csv(r'C:\Users\arif\OneDrive - Higher Education Commission\Teaching\Courses\machine learning\Spring 25\Lab for students\4_AEP_Introducing_holidays.csv', index=False)
print("Final dataset saved successfully!")
```
