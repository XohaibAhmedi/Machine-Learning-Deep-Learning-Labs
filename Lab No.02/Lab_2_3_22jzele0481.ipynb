## 1. Import Matplotlib

```python
import matplotlib.pyplot as plt
```

## 2. Basic Line Plot

```python
# Simple line plot
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 11]

plt.plot(x, y)
plt.show()
```

## 3. Titles, Labels and Grid

```python
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 11]

plt.plot(x, y)

plt.title("Student Marks Over Days")
plt.xlabel("Days")
plt.ylabel("Marks")

plt.grid(True)

plt.show()
```

## 4. Styling: Color, Marker, Line

```python
x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 11]

plt.plot(x, y, color='red', marker='o', linestyle='--', linewidth=2, markersize=8)

plt.title("Styled Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
```

## 5. Subplots

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1
axes[0, 0].plot([1,2,3,4], [10,20,25,30], 'b-o')
axes[0, 0].set_title("Line Plot")

# Plot 2
axes[0, 1].bar(['A','B','C'], [5,7,3], color='orange')
axes[0, 1].set_title("Bar Plot")

# Plot 3
axes[1, 0].scatter([1,2,3,4,5], [2,5,3,8,1], color='green')
axes[1, 0].set_title("Scatter Plot")

# Plot 4
axes[1, 1].hist([1,2,2,3,3,3,4,5,5], bins=5, color='purple')
axes[1, 1].set_title("Histogram")

plt.tight_layout()
plt.show()
```
